```python
games['genres'] = games['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
games['platforms'] = games['platforms'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=100, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[26])), reverse=True, key=lambda x:x[1])
```

```python
list(enumerate(similarity[0]))
```

```python
def recommend(game):
    game_index = new_df[new_df['title'] == game].index[0]
    distances = similarity[game_index]
    games_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[0:5]
    
    for i in games_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Pixel Gear')
```
